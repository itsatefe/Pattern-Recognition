# KNN


1. **Effect of the "K" Value:**
   - A smaller value of "K" (e.g., K=1 or K=3) means that only a few nearest neighbors will be taken into account. This can lead to predictions that are sensitive to local variations in the data, potentially resulting in a noisy prediction.
   - A larger value of "K" (e.g., K=10 or K=20) means that more neighbors are considered, providing a more stable prediction that is less sensitive to individual data points. However, it may result in a biased prediction if there is an imbalance in the classes.

2. **Choosing the Right "K" Value:**

   - A small "K" value tends to make the model more sensitive to noise and outliers, which can lead to overfitting. It's suitable for complex, noisy data with many data points.
   - A larger "K" value provides a smoother decision boundary and is less prone to overfitting but may not capture local patterns well. It's suitable for simpler, well-behaved datasets with less noise.


## 1d data

```python
K = [1, 3, 5, 7]
knn1 = KNN(dataset1,K)
knn1.plot_1d_knn()
```

```python
K = [5, 50, 100, 200]
knn2 = KNN(dataset2,K)
knn2.plot_2d_knn()
```

# Gaussian Kernel

2. **Effect of Bandwidth ():**
   - The bandwidth parameter,  (sigma), controls the width of the Gaussian kernel placed at each data point.
   - Smaller Bandwidth (): A smaller bandwidth results in narrower kernels. Each kernel has a higher peak and covers a smaller region around each data point. This means that the estimated density will be more sensitive to individual data points and capture finer details in the data. Smaller bandwidths lead to higher variability in the density estimate.

   - Larger Bandwidth (): Conversely, a larger bandwidth results in wider kernels. Each kernel has a lower peak and covers a broader region around each data point. This leads to a smoother density estimate that is less sensitive to individual data points. Larger bandwidths lead to lower variability in the density estimate.

3. **Bandwidth Selection:**
   - Selecting an appropriate bandwidth is a critical step in Gaussian kernel density estimation.
   - Too small a bandwidth can lead to overfitting, where the density estimate captures noise in the data and results in a jagged or spiky density curve.
   - Too large a bandwidth can result in oversmoothing, where important features of the data distribution are lost, and the density estimate appears overly smooth.

6. **Interpretation:**
   - The choice of bandwidth affects how you interpret the estimated density. A smaller bandwidth emphasizes local variations and outliers, while a larger bandwidth provides a more global view of the data distribution.


## 1d data

```python
bandwidths = [0.1, 0.3, 0.5, 0.7]
gk1 = GaussianKernel(dataset1.values,bandwidths)
gk1.plot_1d_kde()
```

## 2d data

```python
bandwidths = [0.1, 0.3, 0.5, 0.7]
gk2 = GaussianKernel(dataset2.values,bandwidths)
gk2.plot_2d_kde()
```

# Parzen Window

2. **Effect of Bandwidth ():**
   - The bandwidth parameter,  (sigma), controls the width of the window or kernel function placed around each data point.
   - Smaller Bandwidth (): A smaller bandwidth results in narrower windows. Each window has a higher peak and covers a smaller region around each data point. This means that the estimated density will be more sensitive to individual data points and capture finer details in the data. Smaller bandwidths lead to higher variability in the density estimate.

   - Larger Bandwidth (): Conversely, a larger bandwidth results in wider windows. Each window has a lower peak and covers a broader region around each data point. This leads to a smoother density estimate that is less sensitive to individual data points. Larger bandwidths lead to lower variability in the density estimate.

3. **Bandwidth Selection:**
   - Selecting an appropriate bandwidth is a critical step in Parzen window density estimation.
   - Too small a bandwidth can lead to overfitting, where the density estimate captures noise in the data and results in a jagged or spiky density curve.
   - Too large a bandwidth can result in oversmoothing, where important features of the data distribution are lost, and the density estimate appears overly smooth.

3. **Interpretation:**
   - The choice of bandwidth affects how you interpret the estimated density. A smaller bandwidth emphasizes local variations and outliers, while a larger bandwidth provides a more global view of the data distribution.


```python
bandwidths = [0.1, 0.3, 0.5, 0.7]
pw1 = ParzenWindow(dataset1.values, bandwidths)
pw1.plot_1d_parzen()
pw1.plot_1d_parzen_hist()
```
